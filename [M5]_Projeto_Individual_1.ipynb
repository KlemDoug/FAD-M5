Ambos os gráficos definem no plano cartesiano que a posição de cada categoria é definida pela junção do **nível de impacto** da mesma junto de sua **frequência** no ambiente de trabalho.

> O primeiro gráfico dispõe as categorias de acordo com o quantitativo de amostra oriundo do **setor de Recursos Humanos**, enquanto que o segundo se pauta no quantitativo de **pesquisas de opinião**.

> Algumas conclusões podem ser feitas:

> * Independente do nível de impacto em cada motivação para saída de pessoal, a grande maioria das categorias pontuaram **frequência baixa** para tal decisão;
> * Divergências no posicionamento de algumas categorias podem apontar devolutivas diferentes ou até mesmo enviesadas ao serem feitas por (ex-)funcionários para o setor de Recursos Humanos ou para pesquisas de opinião (vide a categoria '**Quantidade de Tarefas**').

> Essa forma entregue inicialmente é, em verdade, complicada de elucidar e confusa na sua construção. Gráficos de bolhas, apesar de serem ***muito bons para comparar dados categóricos***, possuem um tamanho-limite para o recebimento de dados muito pequeno. Caso esse limite seja ultrapassado, o gráfico se torna difícil para se fazer uma leitura objetiva. É um dilema parecido com o que os amantes dos gráficos tipo **Pizza** acabam enfrentando.

> Outro ponto bem frágil na escolha desse tipo de visualização é como utilizar o **tamanho das bolhas**: o ideal é que a dimensão dos círculos tenha relação direta com sua área, não com outros fatores numéricos que podem ser manipulados exponencialmente. Isso pode acarretar numa má interpretação pelos olhos humanos.

> Uma possível forma de remediar isso **sem alterar o tipo de visualização e nem desperdiçar o trabalho feito** é munindo-se da interatividade que sua ferramenta de *dashboard* fornece: clicando-se ou selecionando bolhas em específico nas seções de painéis e/ou *storyboards* para uma leitura limpa ou destacando informações que estejam mais escondidas.

> Por exemplo: em cada frame do *storyboard*, pode-se dar destaque até, no máximo, 3 categorias (as mais frequentes e/ou as menos frequentes), comparando-as.

> Outras maneiras de trabalhar esses dados de formas distintas da original são esboçadas abaixo:

HEATMAP

image.png

> **Vantagens:** dá pra analisar a relação entre duas ou mais variáveis categóricas, ajudando a identificar associações ou discrepâncias entre elas.

> **Desvantagens:** a comparação é feita pelos valores quantitativos das categorias a partir das outras colunas, ignorando os níveis de impacto que vieram associados a elas.

BAR-IN-BAR

image.png

> **Vantagens:** é prático comparar valores atribuídos a uma mesma variável/categoria vindos de fontes diferentes. Tendo dados temporais, inclusive, isso permite acompanhar a evolução proporcional dessas fontes num intervalo de tempo específico.

> **Desvantagens:** para uma quantidade maior de categorias, a chance de se criar uma poluição visual da mesma magnitude do gráfico de bolhas é alta, pois o tamanho das barras interna e externa fica comprometido.

BARRAS AGRUPADAS

image.png

> **Vantagens:** a comparação entre os valores de cada categoria vindos de diferentes fontes é bem direta, pois as barras estão lado-a-lado, facilitando a leitura visual. Pra múltiplas séries de dados são muito úteis.

> **Desvantagens:** a comparação fica comprometida se o número de séries e subcategorias for muito grande, afinando o espaço de cada barra no grafico e comprometendo muito a interpretação visual.

> De todas as variações apresentadas, a que melhor permite a visualização dos dados é pelo gráfico de **barras agrupadas**: ali é possível  mostrar a frequência das categorias, compará-las e, em seguida, complementar com um gráfico de bolhas se necessário para representar o nível de impacto e o peso de influência sem comprometer com poluição visual ou sobrecarga de informações.
